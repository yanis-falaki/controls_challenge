```python
from environment import Environment
from controllers import SimplePIDController, PIDController
```

```python
env = Environment(data_path='./data/00004.csv')
state = env.reset()
controller = SimplePIDController()
controller = PIDController()

while True:
    params = [state[0], state[1], (state[2], state[3], state[4])]
    action = controller.update(*params)
    next_state, cost, done = env.step(action)
    state = next_state

    if done:
        print(env.get_total_cost()[2])
        break
```
